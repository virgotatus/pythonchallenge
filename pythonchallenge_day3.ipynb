### 第四题

python 挑战第四题  网页http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=45439
思路：
1. python访问这个网址
2. 读取里面的nothing is 后面的数字
3. 读完以后，继续访问变更后的网址
4. 重复 1~3
5. 直到出现结果

```python
# python 访问
import urllib.request
original="http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing="
with urllib.request.urlopen('http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=45439',timeout=10) as response:
    html = response.read()
print(html)
print(html[-7:]) # 数字不一定是5位，所有取最后几位不行
```

```python
# 识别is后面的数字就可以了
# 首先要识别is
#html.index('i')
print(html.decode('utf-8').index('is')) # b''代表bytes
#在 Python3 中，bytes 和 str 的互相转换方式是
#str.encode('utf-8')
#bytes.decode('utf-8')
#https://www.cnblogs.com/walo/p/10608436.html
print(len(html))
print(html[73:])
```

```python
# 获取is后面的数字  函数
def find_num_after_is(html_str):
    idx = html.decode('utf-8').index('is') # example:idx=73
    return html_str[idx+3:].decode('utf-8')
print(find_num_after_is(html))
```

```python
print(original+find_num_after_is(html)) 
# 完成了第2步
```

```python
with urllib.request.urlopen(original+find_num_after_is(html)) as response:
    html = response.read()
print(html)
```

```python
with urllib.request.urlopen(original+find_num_after_is(html)) as response:
    html = response.read()
print(html)
```

```python
html = b"http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=32490"
for i in range(400):
    result = find_num_after_is(html)
    if result == '':
        print(html)
        break
    try:
        with urllib.request.urlopen(original+result,timeout=5) as response:
            html = response.read()
    except:
        pass
    print(i,html)
print(html)
```
