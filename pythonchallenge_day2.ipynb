```python
def judge_window(input_str):
    # input_str = "aAAAaBBBa"  size=7
    flag = True
    for i in range(9):
        if not (i==4 or i==0 or i==8):
            flag = flag and (input_str[i]>='A' and input_str[i]<='Z')
        else:
            flag = flag and (input_str[i]>='a' and input_str[i]<='z')
        if not flag:#flag = False
            return False
    return True

print(judge_window("aAAAaBBBa"))
```
